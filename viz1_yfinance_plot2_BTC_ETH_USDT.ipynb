```python
stock_name=['BTC-USD','ETH-USD','DX=F']
```

```python
i=0
stock=yfinance.Ticker(stock_name[i])
df1=stock.history(period='1y')
df1.head()
```

```python
i=1
stock=yfinance.Ticker(stock_name[i])
df2=stock.history(period='1y')
df2.head()
```

```python
i=2
stock=yfinance.Ticker(stock_name[i])
df3=stock.history(period='1y')
df3.head()
```

```python
df1.Close.plot(figsize=(15,3))
```

```python
df2.Close.plot(figsize=(15,3))
```

```python
df3.Close.plot(figsize=(15,3))
```

```python
import pandas as pd
df_many = pd.concat([df1.Close, df2.Close,df3.Close], axis=1)
df_many.columns=['BTC-USD','ETH-USD','DX=F']
df_many.head()
```

```python
# Plot the grid lines
import matplotlib.pyplot as plt

((df_many.pct_change()+1).cumprod()).plot(figsize=(10, 7))
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```
