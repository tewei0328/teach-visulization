```python
import plotly
import pandas as pd
# Load data
df = pd.read_csv(
    "https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv")
df.columns = [col.replace("AAPL.", "") for col in df.columns]
```

```python
df.head()
```

```python
import plotly.graph_objects as go
line1 = go.Scatter(x=df['Date'], y=df['Close'], name='Close')
fig  = go.Figure(line1)
fig.show()
```

```python
import plotly.graph_objects as go
line1 = go.Scatter(x=df['Date'], y=df['Close'], name='Close')
line2 = go.Scatter(x=df['Date'], y=df['Low'], name='Low')
fig  = go.Figure([line1, line2])
fig.update_layout(
    title = "Apple Stock Price",
    xaxis_title = "Date",
    yaxis_title = "Price"
)
fig.show()
```
