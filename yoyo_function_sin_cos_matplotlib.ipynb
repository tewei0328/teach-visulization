```python
import numpy as np 
import matplotlib.pyplot as plt 
X = np.linspace(0, 2 * np.pi, 100) 
YSinValues = np.sin(X) 
YCosValues = np.cos(X) 
plt.plot(X, YSinValues) 
plt.plot(X, YCosValues) 
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
X = np.linspace(0, 2 * np.pi, 100) 
YSinValues = np.sin(X)*np.sin(X) 
YCosValues = -np.cos(X) 
plt.plot(X, YSinValues) 
plt.plot(X, YCosValues) 
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
X = np.linspace(0, 2 * np.pi, 100) 
Y1 = np.sin(X) 
Y2 = np.sin(X)*np.sin(X) 
Y3 = -np.cos(X)
Y = np.sin(X)*np.sin(X) -np.cos(X) 
plt.plot(X, Y) 
plt.plot(X, Y1)
plt.plot(X, Y2)
plt.plot(X, Y3)   
plt.show()
```
